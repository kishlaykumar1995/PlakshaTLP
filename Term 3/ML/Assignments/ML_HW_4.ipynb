```python
#Importing the required packages
import random
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
# Constants for XOR operation
#Inputs for XOR
XOR_INPUTS = [[0, 0], [0, 1], [1, 0], [1, 1]]

#Output for XOR
XOR_OUTPUTS = [[0], [1], [1], [0]]
```

```python
#Weights as given in the class
w11 = 3.7
w21 = 3.7
b1 = -1.5

w12 = 2.9
w22 = 2.9
b2 = -4.5


o1 = 4.5
o2 = -5.2
ob = -2

#Sigmoid function
def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

# x already sigmoided
def sigmoid_prime(x): 
    return x * (1 - x)


def predict(i1,i2):    
    s1 = w11 * i1 + w21 * i2 + b1
    s1 = sigmoid(s1)
    s2 = w12 * i1 + w22 * i2 + b2
    s2 = sigmoid(s2)
    
    output = s1 * o1 + s2 * o2 + ob
    output = sigmoid(output)
    
    return output

'''The learn function is used to update the weights and biases of the network. It takes in the input values, the target output, and an optional learning rate parameter. 
It first calculates the output of the hidden units and the output of the network using the predict function. Then, it calculates the error between the target output and the 
predicted output. The error is used to calculate the derivatives of the error with respect to the weights and biases using the sigmoid_prime function. Finally, it updates 
the weights and biases using these derivatives and the learning rate.'''

def learn(i1,i2,target, alpha=0.2):
    global w11,w21,b1,w12,w22,b2
    global o1,o2,ob
    
    s1 = w11 * i1 + w21 * i2 + b1
    s1 = sigmoid(s1)
    s2 = w12 * i1 + w22 * i2 + b2
    s2 = sigmoid(s2)
    
    output = s1 * o1 + s2 * o2 + ob
    output = sigmoid(output)

    print("for input", i1,i2, "expected", target, "predicted", f"{output:4.4}", "which is", "correct" if round(output)==target else "incorrect")
    
    error = target - output
    derror = error * sigmoid_prime(output)
    
    ds1 = derror * o1 * sigmoid_prime(s1)
    ds2 = derror * o2 * sigmoid_prime(s2)
    
    o1 += alpha * s1 * derror
    o2 += alpha * s2 * derror
    ob += alpha * derror
    
    #Weight update
    w11 += alpha * i1 * ds1
    print("W11:", w11)
    w21 += alpha * i2 * ds1
    print("W21:", w21)
    b1 += alpha * ds1
    print("b1:", b1)
    w12 += alpha * i1 * ds2
    print("W12:", w12)
    w22 += alpha * i2 * ds2
    print("W22:", w22)
    b2 += alpha * ds2
    print("b2:", b2)

    print("Hidden1:", o1)
    print("Hidden2:", o2)
    print("Basis to Output:", ob, end = '\n\n')
```

```python
print("Initial W11(given):", w11)
print("Initial W21(given):", w21)
print("Initial b1(given):", b1)
print("Initial W12(given):", w12)
print("Initial W22(given):", w22)
print("Initial b2(given):", b2)
print("Initial Hidden1(given):", o1)
print("Initial Hidden2(given):", o2)
print("Initial Basis to Output(given):", ob, end = '\n\n')

# Train the network for 10000 epochs
for epoch in range(1, 3):
    # Shuffle the input examples
    indexes = [0, 1, 2, 3]
    # random.shuffle(indexes)

    print(f"Updated weights and biases after {epoch} iteration:", end = '\n\n')
    # Train the network on each example
    for j in indexes:
        learn(XOR_INPUTS[j][0],XOR_INPUTS[j][1],XOR_OUTPUTS[j][0], alpha=0.2)
    
    # Calculate mean squared error
    cost = 0
    for j in range(4):
        output = predict(XOR_INPUTS[j][0],XOR_INPUTS[j][1])
        cost += (XOR_OUTPUTS[j][0] - output) ** 2
    cost /= 4
    print("epoch", epoch, "mean squared error:", cost, end = '\n\n')
```
