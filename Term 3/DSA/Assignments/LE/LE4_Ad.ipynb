```python
# Lab Exercise - 4 
# Priority Queues

import sys

class PriorityQueue:

    def __init__(self, maxsize):
      self.maxsize = maxsize
      self.size = 0
      # creating the list with indexing starting from 1 for simplicity
      self.Heap = [0]*(self.maxsize + 1)
      self.Heap[0] = -1 * sys.maxsize
      self.FRONT = 1

    # minHeapify method to minHeapify the node at pos
    def minHeapify(self, pos):
        parentindex=int(pos/2)
        if self.isPqEmpty():
            return
        elif pos<2 or parentindex<1:
            return
        else:
            if self.Heap[pos]<self.Heap[parentindex]:
                self.Heap[pos],self.Heap[parentindex]=self.Heap[parentindex],self.Heap[pos]
                self.minHeapify(parentindex)
            else:
                return
                       
            
            
      

    # Write this function to insert a node into the heap

    def insert(self, element):
        self.size+=1
        if self.size>self.maxsize:
            return
        else:
            self.Heap[self.size]=element
            self.minHeapify(self.size)
            
      

    # Write this function to delete the rootNode
    def delete(self):
        if self.isPqEmpty():
            return
        else:
            self.Heap[self.FRONT]=self.Heap[self.size]
            self.Heap[self.size]=0
            self.size-=1
            if not self.isPqEmpty():
                start=1
                while (start*2)<self.size:
                    if self.Heap[(start*2)+1]==0:
                        if self.Heap[start]>self.Heap[start*2]:
                            self.Heap[start],self.Heap[start*2]=self.Heap[start*2],self.Heap[start]
                        else:
                            return
                    else:
                        left=self.Heap[start*2]
                        right=self.Heap[(start*2)+1]
                        if left<=right:
                            self.Heap[start],self.Heap[start*2]=self.Heap[start*2],self.Heap[start]
                            start=start*2
                        else:
                            self.Heap[start],self.Heap[(start*2)+1]=self.Heap[(start*2)+1],self.Heap[start]
                            start=(start*2)+1
                            
                            
                        
      

    # Write this function to return the rootNode (here the minimum element in PQ)
    def minimumElement(self):
        if self.isPqEmpty():
            return None
        else:
            return self.Heap[self.FRONT]
      

    # Write this function to return the size of the PriorityQueue
    def sizeOfPq(self):
        return self.size
      

    # Write this function to return if the priorityQueue is empty or not
    # Return boolean value
    def isPqEmpty(self):
        if self.size==0:
            return True
        else:
            return False
    

    # Function to print the contents of the heap
    def printPriorityQueue(self):
        for i in range(1, (self.size//2)+1):
            print("Parent : "+ str(self.Heap[i])+" Left Child : "+
                      str(self.Heap[2 * i])+" Right Child : "+
                      str(self.Heap[2 * i + 1]))

    # Function to build the PriorityQueue using minHeap
    def priorityQueue(self):
        for pos in range(self.size//2, 0, -1):
            self.minHeapify(pos)
```
