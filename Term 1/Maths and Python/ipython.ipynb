```python
def find_minima():  
    alpha=-100
    d_alpha_1=50*((alpha)**2)-142*(alpha)+101
    toggle,decrement = 0,1.0

    while True:
        if toggle==0:
            alpha=alpha+decrement
        elif toggle==1:
            alpha=alpha-decrement
        
        d_alpha_2=50*((alpha)**2)-142*(alpha)+101

        if d_alpha_1 < d_alpha_2:
            if toggle==0:
                toggle=1
            else:
                toggle=0
            decrement=decrement/2
            
        if decrement < 0.0001:
            break

        d_alpha_1 = d_alpha_2

    print(alpha)

find_minima()
```

```python
def alpha_min():
    alpha=0
    drop=1
    while True:
        val1=50*(alpha**2)-142*alpha+101
        val2=50*((alpha+drop)**2)-142*(alpha+drop)+101

        if val1<val2:
            alpha=alpha-2*drop
            drop=drop/10
        
        if drop == 0.0001:
            break
        alpha+=drop

    print(alpha)

alpha_min()
```

```python
# Hilbert curve
# Str = ri, ci, rd;  Left = ci, ri, cd; Right = cd, ri, ci

def get_left(v):
    ls = []
    r=v[0];c=v[1]
    ls.append((r,c))
    c=c+1
    ls.append((r,c))
    r=r+1
    ls.append((r,c))
    c=c-1
    ls.append((r,c))
    return ls

def get_right(v):
    ls = []
    r=v[0];c=v[1]
    ls.append((r,c))
    c=c-1
    ls.append((r,c))
    r=r-1
    ls.append((r,c))
    c=c+1
    ls.append((r,c))
    return ls

def get_str(v):
    ls = []
    r=v[0];c=v[1]
    ls.append((r,c))
    r=r+1
    ls.append((r,c))
    c=c+1
    ls.append((r,c))
    r=r-1
    ls.append((r,c))
    return ls

def get_inv(v):
    ls = []
    r=v[0];c=v[1]
    ls.append((r,c))
    r=r-1
    ls.append((r,c))
    c=c+1
    ls.append((r,c))
    r=r+1
    ls.append((r,c))
    return ls
```

```python
def vertex_find(v, dir):
    if dir == 'R':
        return ((v[0],v[1]+1))
    elif dir == 'L':
        return ((v[0],v[1]-1))
    elif dir == 'U':
        return ((v[0]-1,v[1]))
    elif dir == 'D':
        return ((v[0]+1,v[1]))
    
def anticlock(unit):
    unit.reverse()
    l=[]
    for i in range(0,len(unit)):
        if unit[i]=="R":
            l.append("D")
        elif unit[i]=="D":
            l.append("L")
        elif unit[i]=="L":
            l.append("U")
        elif unit[i]=="U":
            l.append("R")
    unit.reverse()
    return l
    

def clock(unit):
    unit.reverse()
    l=[]
    for i in range(len(unit)):
        if unit[i]=="U":
            l.append("L")
        elif unit[i]=="R":
            l.append("U")
        elif unit[i]=="D":
            l.append("R")
        elif unit[i]=="L":
            l.append("D")
    unit.reverse()
    return l
    
def get_pattern(left_pattern,right_pattern, unit):
    pattern=[]
    for i in range(len(left_pattern)):
        pattern.append(left_pattern[i])
    pattern.append("D")
    for i in range(len(unit)):
        pattern.append(unit[i])
    pattern.append("R")
    for i in range(len(unit)):
        pattern.append(unit[i])
    pattern.append("U")
    for i in range(len(right_pattern)):
        pattern.append(right_pattern[i])
    return pattern

def next_pattern():
    unit=['D','R','U']
    count=0
    while True:
        #print("Next pattern")
        #print(get_pattern(anticlock(unit),clock(unit),unit))
        unit=get_pattern(anticlock(unit),clock(unit),unit)
        z = [(1,1)]
        for dir in unit:
            z.append(vertex_find(z[-1],dir))
        print(z)
        count=count+1 
        if count==5:
            break
        print("\n")
    
next_pattern()
```

```python
def binary_exponent(a,b):
    """
    This function performs binary exponentiation iteratively
    """
    a_pow=a                  # This variable stores the power of 'a' in a particular iteration

    #1st iteration - Here the power of a is 1
    if b%2==1:
        result = a
    else:
        result = 1
    b = b//2

    while b>0:
        rem = b%2
        b = b//2
        a_pow = a_pow*a_pow  # Power of a increases in powers of 2 in every iteration
        if rem==1:           # if remainder is 1, then multiply the result with current iteration of a_pow 
            result = result*a_pow


    return result
binary_exponent(6,12)
```
